Install psycopg2 by running the following command in your terminal or command prompt:

```python
# Import libraries
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import text

# Create engine: engine
# engine = create_engine('postgresql://username:password@localhost:5432/database_name')
engine = create_engine('postgresql://mushasho:perrenque@localhost:5432/STAFFING_ABROAD')

# Create a connection object: conn
conn = engine.connect()

# Execute query and store records in DataFrame: df
df = pd.read_sql_query('SELECT * FROM merchant_table', conn)

# Close connection
conn.close()

# Print head of DataFrame
print(df.head())
```

```python
# 5. Relaciones entre las variables 

Features_Table = otu_taxa_ws.iloc[:, 0:22]
Taxonomic_table = otu_taxa_ws.iloc[:, 22:]
OTU = otu_taxa_ws.iloc[:, 22]

print(OTU.head(5))
print(OTU.head(5))

Features_Table.corr()
sns.heatmap(Features_Table.corr())
```

```python
# 3. Data Preparation

# Relative abundance of a taxa in a sample

from sklearn import preprocessing

# Define axis used to normalize the data along. If 1, independently normalize each sample
Features_Table_norm = preprocessing.normalize(Features_Table, axis=1)

Relative_abundance = pd.DataFrame(Features_Table_norm,columns=[Features_Table.columns])

print('Original Data \n',Features_Table.head(2))
print('Normalized Data \n',Relative_abundance.head(2))

sns.heatmap(Relative_abundance.corr())
```

```python
sns.catplot(y="Phylum", data=otu_taxa_ws,
            kind="count",
            col="Kingdom")
            
# Show plot
plt.show()
```

```python
# Joins

Features_Table_OTU = Features_Table['OTU'] = OTU
#print(Features_Table.head(5))

Relative_abundance_OTU = Relative_abundance['OTU'] = OTU
#print(Taxonomic_table.head(5))

Relative_abundance.to_clipboard(index=False)
Relative_abundanceclip=pd.read_clipboard(sep='\t')

Taxonomic_table.to_clipboard(index=False)
Taxonomic_tableclip=pd.read_clipboard(sep='\t')

otu_taxa_ws_ra = pd.merge(Relative_abundanceclip,Taxonomic_tableclip,on='OTU',how='outer')

#otu_taxa_ws_ra = Relative_abundance.merge(Taxonomic_table, on='OTU', how='outer')
#print(otu_taxa_ws_ra.head(5))
```

```python
# Melt on abundance otu_taxa_ws

id_vars = ["OTU",
           "Kingdom",
           "Phylum",
           "Class",
           "Order",
           "Family",
           "Genus",
           "Species",
           "SciName"]

formatted_otu_taxa_ws = pd.melt(frame=otu_taxa_ws,
                                id_vars=id_vars,
                                var_name="SampleID",
                                value_name="abundance")
formatted_otu_taxa_ws = formatted_otu_taxa_ws.sort_values(by=["OTU"])                       
formatted_otu_taxa_ws.head(10)
```

```python
# Melt on abundance otu_taxa_ws_ra

id_vars = ["OTU",
           "Kingdom",
           "Phylum",
           "Class",
           "Order",
           "Family",
           "Genus",
           "Species",
           "SciName"]

formatted_otu_taxa_ws_ra = pd.melt(frame=otu_taxa_ws_ra,
                                id_vars=id_vars,
                                var_name="SampleID",
                                value_name="abundance")
formatted_otu_taxa_ws_ra = formatted_otu_taxa_ws_ra.sort_values(by=["OTU"])                       
formatted_otu_taxa_ws_ra.head(10)
```

```python
# Join with metadata

abundance_metadata = pd.merge(metadata,formatted_otu_taxa_ws,on='SampleID',how='inner')
abundance_metadata_ra = pd.merge(metadata,formatted_otu_taxa_ws_ra,on='SampleID',how='inner')

#print(abundance_metadata_ra.head(5))
```

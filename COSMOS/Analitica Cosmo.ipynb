3. Data preparation & Data cleaning

```python
df['fecha_hecho'] =  pd.to_datetime(df['fecha_hecho'], format = "%Y-%m-%d %H:%M:%S")
df['fecha_ingestion'] =  pd.to_datetime(df['fecha_ingestion'], format = "%Y-%m-%d %H:%M:%S")
display(df.info()) # view information about the dataset, such as column types and non-null values
```

4. Exploratory Data Analysis

```python
display(df.describe()) # view descriptive statistics of the dataset
```

```python
categorical_data = df.select_dtypes(include=['object']).columns.tolist()

for i in categorical_data:
    if len(df[i].unique()) > 1:
        val_i = df[i].value_counts()
        print('\033[1m' + str(i).upper() + '\033[0m'+ "\n" + val_i.to_string() + "\n")
```

```python
for i in categorical_data:
    if len(df[i].unique()) > 1:
        top_categories = df[i].value_counts().nlargest(10).index.tolist()
        df[i] = df[i].replace(set(df[i].unique()) - set(top_categories), 'otros')
        plot_i = df[i].value_counts().plot(kind='bar')
        plot_i.set_xticklabels(plot_i.get_xticklabels(), rotation=45, ha="right")
        plt.title(i)
        plt.show()
```
